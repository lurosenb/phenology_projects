```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from PIL import Image
```

## Onetime

```python
onetime_train = pd.read_csv('../datasets/onetime-train.csv')
onetime_test = pd.read_csv('../datasets/onetime-test.csv')
onetime = pd.concat([onetime_train, onetime_test])
onetime = gpd.GeoDataFrame(onetime, geometry=gpd.points_from_xy(onetime.Longitude, onetime.Latitude))
onetime = onetime[onetime.Longitude>-112]
onetime = onetime[onetime.Latitude<33]
```

```python
fig_size = (1, 3)
f, ax = plt.subplots(1, figsize=fig_size)
ax = onetime.plot(ax=ax,markersize=0.1,c='red',marker="2",figsize=fig_size)
ax.set_axis_off()
ax.apply_aspect()
ax.figure.savefig('onetime-distribution.png', dpi=600, bbox_inches='tight')
```

```python
overlay = Image.open("onetime-distribution.png").convert("RGBA")
background = Image.open("onetime-satellite-overlay.jpg").convert("RGBA").resize(overlay.size)
new_img = Image.blend(background, overlay, 0.35)
new_img.save("onetime-geo-vis.png","PNG")
new_img
```

## Pheno

```python
pheno_train = pd.read_csv('../datasets/pheno-train.csv')
pheno_test = pd.read_csv('../datasets/pheno-test.csv')
pheno = pd.concat([pheno_train, pheno_test])
pheno = gpd.GeoDataFrame(pheno, geometry=gpd.points_from_xy(pheno.Longitude, pheno.Latitude))
pheno = pheno[pheno.Longitude<-105]
pheno = pheno[pheno.Latitude<33]
pheno = pheno[pheno.Latitude>28]
```

```python
fig_size = (2, 1)
f, ax = plt.subplots(1, figsize=fig_size)
ax = pheno.plot(ax=ax,markersize=0.1,c='red',marker="2",figsize=fig_size)
ax.set_axis_off()
ax.apply_aspect()
ax.figure.savefig('pheno-distribution.png', dpi=600, bbox_inches='tight')
```

```python
overlay = Image.open("pheno-distribution.png").convert("RGBA")
background = Image.open("pheno-satellite-overlay.jpg").convert("RGBA").resize(overlay.size)
new_img = Image.blend(background, overlay, 0.35)
new_img.save("pheno-geo-vis.png","PNG")
new_img
```
